```python
# Using updated dataframe from preprocessing notebook
df = pd.read_csv("../data/listings.csv")
display(df)
```

In ridge regression, our goal is to minimize the value $||y - Xw||^2_2 + \alpha * ||w||^2_2$. We do this with the following code;

```python
# Performing Ridge Regression
# Start with a standard value of alpha
rm = Ridge(
    alpha=0.5
)  # normalize parameter being removed soon in scikit.learn, will avoid

# Lets test with very basic model first
# sklearn is a little picky, so we turn our coulms into 2-D arrays

X, y = (
    np.array(df["number_of_reviews"]).reshape(-1, 1),
    np.array(df["price"]).reshape(-1, 1),
)
rm.fit(X, y)
rm.predict(X)

# Basic for now...

# Let's make a more diverse model
npdf = df.to_numpy()
# neighbourhood_group, neighbourhood, room_type, minimun_nights,
# number_of_reviews, reviews_per_month, calculated_host_listings_count, availability_365
X = npdf[:, [4, 5, 6, 7, 9, 10, 12, 13, 14]]
# price
y = npdf[:, 8]  # price
rm.fit(X, y)
```
